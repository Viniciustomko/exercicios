```python
# 1 Explique, com suas palavras, a diferença entre algoritmo, fluxograma e pseudocódigo

Um algoritmo é a sequência de passos lógicos para resolver um problema. Um fluxograma é a representação visual desse algoritmo, usando símbolos e setas para mostrar o fluxo do processo. Já o pseudocódigo é uma forma textual, que utiliza uma linguagem de alto nível.
```

```python
#4 Modele um algoritmo que calcule o tempo necessário para encher um reservatório de água com base no volume (litros) e na vazão da torneira (litros/minuto).

volume = float(input("Volume do reservatório (L): "))
vazao = float(input("Vazão da torneira (L/min): "))

tempo_min = volume / vazao
horas = int(tempo_min // 60)
minutos = int(tempo_min % 60)
segundos = int((tempo_min - int(tempo_min)) * 60)

print(f"Tempo: {horas}h {minutos}min {segundos}s")
```

```python
#5 Crie um algoritmo que simule o processo de autenticação de usuário, com três tentativas de senha.

usuario_cadastrado = "taz"
senha_cadastrada = "1234"

# tentativas
tentativas_max = 3
tentativas = 0


while tentativas < tentativas_max:
    usuario = input("Digite o nome de usuário: ")
    senha = input("Digite a senha: ")

    if usuario == usuario_cadastrado and senha == senha_cadastrada:
        print("Autenticação bem-sucedida! Bem-vindo(a).")
        break
    else:
        tentativas += 1
        print(f"Usuário ou senha incorretos. Tentativa {tentativas} de {tentativas_max}.")

        if tentativas == tentativas_max:
            print("Número máximo de tentativas atingido. Acesso bloqueado.")
```

```python
# 19. Desenvolva um programa que leia idades até que o usuário digite 0, e calcule a média de idade.

total = 0
cont = 0

while True:
    idade = int(input("Digite a idade (0 para sair): "))
    if idade == 0:
        break
    total += idade
    cont += 1

if cont > 0:
    print("Média das idades:", total / cont)
else:
    print("Nenhuma idade digitada.")
```

```python
# 20. Escreva um algoritmo que leia uma sequência de números e conte quantos são positivos, negativos e zero.

positivos = 0
negativos = 0
zeros = 0

n = int(input("Quantos números você quer digitar? "))


for i in range(n):
    num = float(input(f"Digite o {i+1}º número: "))
    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        zeros += 1


print(f"Números positivos: {positivos}")
print(f"Números negativos: {negativos}")
print(f"Números iguais a zero: {zeros}")

```

```python
# 22  Desenvolva um algoritmo que leia as notas de 4 provas de 5 alunos e calcule a média individual e geral.

num_alunos = 5
num_provas = 4

medias = []

for i in range(num_alunos):
    notas = [float(input(f"Aluno {i+1}, nota {j+1}: ")) for j in range(num_provas)]
    media = sum(notas) / num_provas
    medias.append(media)
    print(f"Média do aluno {i+1}: {media:.2f}")

media_geral = sum(medias) / num_alunos
print(f"Média geral da turma: {media_geral:.2f}")

```

```python
# 23. Faça um algoritmo que leia um vetor com 10 números e exiba apenas os valores maiores que a média.

numeros = []


for i in range(10):
    num = float(input(f"Digite o {i+1}º número: "))
    numeros.append(num)


media = sum(numeros) / len(numeros)

print(f"\nA média dos números é: {media}")


print("Números maiores que a média:")
for num in numeros:
    if num > media:
        print(num)
```

```python
# 24. Crie um algoritmo que leia duas listas de produtos e mostre os produtos que aparecem em ambas.

lista1 = input("Digite os produtos da primeira lista separados por vírgula: ").split(',')


lista2 = input("Digite os produtos da segunda lista separados por vírgula: ").split(',')


lista1 = [produto.strip() for produto in lista1]
lista2 = [produto.strip() for produto in lista2]

# Encontrar produtos em comum usando conjuntos
produtos_comuns = list(set(lista1) & set(lista2))


if produtos_comuns:
    print("Produtos que aparecem em ambas as listas:", produtos_comuns)
else:
    print("Não há produtos em comum entre as listas.")
```

```python
# 25. Desenvolva um algoritmo que leia uma matriz 3x3 e exiba a soma dos elementos da diagonal principal.

matriz = []


for i in range(3):
    linha = []
    for j in range(3):
        valor = int(input(f"Digite o elemento da posição [{i}][{j}]: "))
        linha.append(valor)
    matriz.append(linha)


soma_diagonal = 0
for i in range(3):
    soma_diagonal += matriz[i][i]  # elementos onde linha == coluna


print("A soma dos elementos da diagonal principal é:", soma_diagonal)
```

```python
# 26 Crie uma função que receba um número e retorne seu fatorial.

def fatorial(n):
    if n < 0:
        return "Fatorial não definido para números negativos"
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

# Exemplo de uso
print(fatorial(3))
```

```python
#27. Desenvolva uma função que receba dois números e retorne o máximo divisor comum (MDC).

def mdc(a, b):

    while b != 0:
        a, b = b, a % b
    return a

# Exemplo de uso
num1 = 10
num2 = 12
print(f"O MDC de {num1} e {num2} é {mdc(num1, num2)}")
```

```python
# 29. Crie uma função que receba um número e verifique se ele é perfeito (a soma de seus divisores é igual a ele).
def eh_perfeito(n):
    return n == sum(i for i in range(1, n) if n % i == 0)

# Exemplo de uso
print(eh_perfeito(6))
print(eh_perfeito(12))
```

```python
# 30 Implemente um programa modular com as funções: lerVetor(), maiorElemento(), mediaVetor() e exibirResultados().

def Vetor():
    return [float(input(f"Digite o elemento {i+1}: ")) for i in range(int(input("Digite o tamanho do vetor: ")))]

def maiorElemento(vetor):
    return max(vetor) if vetor else None

def mediaVetor(vetor):
    return sum(vetor)/len(vetor) if vetor else 0

def exibirResultados(vetor, maior, media):
    print("Vetor:", vetor)
    print("Maior elemento:", maior)
    print("Média:", media)

def main():
    vetor1 = Vetor()
    exibirResultados(vetor1, maiorElemento(vetor1), mediaVetor(vetor1))

if __name__ == "__main__":
    main()
```
