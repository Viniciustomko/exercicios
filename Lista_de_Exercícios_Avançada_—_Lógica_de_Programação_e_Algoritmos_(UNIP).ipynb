```python
 #2. Tipos de Dados e Variáveis
 # 1 Crie um algoritmo que leia dados pessoais (nome, idade, sexo, altura, peso) e calcule o IMC, classificando o resultado em categorias (abaixo do peso, normal, sobrepeso, obesidade).

def calcular_imc(peso, altura):
    return peso / (altura ** 2)

# Função para classificar o IMC
def classificar_imc(imc):
    if imc < 18.5:
        return "Abaixo do peso"
    elif 18.5 <= imc < 25:
        return "Normal"
    elif 25 <= imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidade"

# Programa principal
def main():
    # Leitura de dados pessoais
    nome = input("Digite seu nome: ")
    idade = int(input("Digite sua idade: "))
    sexo = input("Digite seu sexo (M/F): ")
    altura = float(input("Digite sua altura: "))
    peso = float(input("Digite seu peso em kg: "))

    # Cálculo do IMC
    imc = calcular_imc(peso, altura)

    # Classificação do IMC
    classificacao = classificar_imc(imc)

    # Exibição dos resultados
    print(f"\nNome: {nome}")
    print(f"Idade: {idade}")
    print(f"Sexo: {sexo}")
    print(f"Altura: {altura} m")
    print(f"Peso: {peso} kg")
    print(f"IMC: {imc:.2f}")
    print(f"Classificação: {classificacao}")

if __name__ == "__main__":
    main()
```

```python
# 2 Escreva um algoritmo que determine o tipo de triângulo (equilátero, isósceles, escaleno) a partir de três lados informados.

def forma_triangulo(a, b, c):
    return a + b > c and a + c > b and b + c > a

# Função para determinar o tipo de triângulo
def tipo_triangulo(a, b, c):
    if a == b == c:
        return "Equilátero"
    elif a == b or b == c or a == c:
        return "Isósceles"
    else:
        return "Escaleno"

# Entrada dos lados
a = float(input("Digite o lado a: "))
b = float(input("Digite o lado b: "))
c = float(input("Digite o lado c: "))

# Verificação e saída
if forma_triangulo(a, b, c):
    print(f"O triângulo é {tipo_triangulo(a, b, c)}.")
else:
    print("Os lados informados não formam um triângulo.")
```

```python
#5. Estruturas de Repetição
# 1 Crie um algoritmo que determine a sequência de Fibonacci até um número N informado pelo usuário.

N = int(input("Digite um número inteiro N: "))


a, b = 0, 1

#sequência
fibonacci = []


while a <= N:
    fibonacci.append(a)
    a, b = b, a + b


print("Sequência de Fibonacci até", N, "é:")
print(fibonacci)
```

```python
#2 Desenvolva um algoritmo que calcule o fatorial de um número usando tanto o laço PARA quanto o laço ENQUANTO, mostrando a diferença entre as abordagens.

#Usando for (laço determinado)

numero = int(input("Digite um número para calcular o fatorial: "))
fatorial_for = 1

for i in range(1, numero + 1):
    fatorial_for *= i  # Multiplica o acumulador pelo valor atual

print(f"Fatorial de {numero} usando FOR é: {fatorial_for}")

#Usando while (laço indeterminado)

numero = int(input("Digite um número para calcular o fatorial: "))
fatorial_while = 1
contador = 1

while contador <= numero:
    fatorial_while *= contador
    contador += 1

print(f"Fatorial de {numero} usando WHILE é: {fatorial_while}")
```

```python
#  2 Faça um algoritmo de menu interativo (usando CASE) com as seguintes opções: calcular área de figuras geométricas (quadrado, triângulo, círculo, retângulo).

import math

def menu():

    print("1 - Quadrado")
    print("2 - Retângulo")
    print("3 - Triângulo")
    print("4 - Círculo")
    print("0 - Sair")
    return int(input("Escolha uma opção: "))

while True:
    opcao = menu()

    match opcao:
        case 1:
            lado = float(input("Digite o lado do quadrado: "))
            area = lado ** 2
            print(f"Área do quadrado: {area:.2f}")
        case 2:
            base = float(input("Digite a base do retângulo: "))
            altura = float(input("Digite a altura do retângulo: "))
            area = base * altura
            print(f"Área do retângulo: {area:.2f}")
        case 3:
            base = float(input("Digite a base do triângulo: "))
            altura = float(input("Digite a altura do triângulo: "))
            area = (base * altura) / 2
            print(f"Área do triângulo: {area:.2f}")
        case 4:
            raio = float(input("Digite o raio do círculo: "))
            area = math.pi * raio ** 2
            print(f"Área do círculo: {area:.2f}")
        case 0:
            print("Encerrando o programa... Até mais!")
            break
        case _:
            print("inválido")
```

```python
# 7. Estruturas de Repetição Avançadas
# 1 Escreva um algoritmo que leia diversos números e informe: a média dos pares; a média dos
# ímpares; o maior e o menor número lido. A entrada termina quando o usuário digitar um número negativo.

soma_pares = 0
cont_pares = 0
soma_impares = 0
cont_impares = 0
maior = None
menor = None

while True:
    num = int(input("Digite um número (negativo para encerrar): "))

    if num < 0:
        break


    if maior is None or num > maior:
        maior = num
    if menor is None or num < menor:
        menor = num


    if num % 2 == 0:
        soma_pares += num
        cont_pares += 1
    else:
        soma_impares += num
        cont_impares += 1


media_pares = soma_pares / cont_pares if cont_pares > 0 else 0
media_impares = soma_impares / cont_impares if cont_impares > 0 else 0



print(f"Média dos números pares: {media_pares:.2f}")
print(f"Média dos números ímpares: {media_impares:.2f}")
print(f"Maior número: {maior}")
print(f"Menor número: {menor}")

```

```python
# 2 Faça um algoritmo que simule um sistema de votação, com três candidatos e opção de voto nulo. O programa deve contar os votos e exibir o vencedor ao final.

def sistema_votacao():

    votos = {
        "Candidato 1": 0,
        "Candidato 2": 0,
        "Candidato 3": 0,
        "Nulo": 0
    }



    print("1 - Candidato 1")
    print("2 - Candidato 2")
    print("3 - Candidato 3")
    print("0 - Voto Nulo")
    print("9 - Encerrar votação")

    while True:
        voto = input("Seu voto: ")

        if voto == "9":
            break
        elif voto == "1":
            votos["Candidato 1"] += 1
        elif voto == "2":
            votos["Candidato 2"] += 1
        elif voto == "3":
            votos["Candidato 3"] += 1
        elif voto == "0":
            votos["Nulo"] += 1
        else:
            print("Opção inválida")
            continue


    for candidato, quantidade in votos.items():
        print(f"{candidato}: {quantidade} voto(s)")

    # Determina o vencedor
    vencedor = max(votos, key=votos.get)
    # Verifica empate
    valores = list(votos.values())
    if valores.count(max(valores)) > 1:
        print("Houve empate!")
    else:
        print(f"\nO vencedor é: {vencedor}!")


sistema_votacao()
```

```python
 #9. Funções
 #1 Implemente uma função que receba dois números e retorne o máximo divisor comum (MDC) entre eles.

def mdc(a, b):
    while b != 0:
        a, b = b, a % b
    return a

num1 = 48
num2 = 18
print(f"O MDC de {num1} e {num2} é {mdc(num1, num2)}")
```

```python
#2 Crie uma função que receba um vetor e retorne o maior elemento, o menor elemento e a média.

def analisar_vetor(vetor):
    if not vetor:  # Verifica se a lista não está vazia
        return None, None, None

    maior = max(vetor)
    menor = min(vetor)
    media = sum(vetor) / len(vetor)

    return maior, menor, media

# Exemplo de uso:
numeros = [ 8, 23, 4]
maior, menor, media = analisar_vetor(numeros)
print(f"Maior: {maior}, Menor: {menor}, Média: {media}")
```

```python
# 4 Implemente um programa modular que contenha: uma função para ler um vetor; uma função para somar os elementos pares; e uma função para imprimir os resultados.

def ler_vetor():
    vetor = []
    n = int(input("Quantos elementos terá o vetor? "))
    for i in range(n):
        valor = int(input(f"Digite o elemento {i+1}: "))
        vetor.append(valor)
    return vetor


def soma_pares(vetor):
    soma = 0
    for num in vetor:
        if num % 2 == 0:
            soma += num
    return soma


def imprimir_resultado(soma):
    print(f"A soma dos elementos pares é: {soma}")


def main():
    vetor = ler_vetor()
    soma = soma_pares(vetor)
    imprimir_resultado(soma)

# Executa o programa
if __name__ == "__main__":
    main()
```
