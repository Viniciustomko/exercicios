```python
#12. Crie um algoritmo que leia o salário de um funcionário e calcule o novo valor com base na regra:
#     - até R$ 1.500 → aumento de 15% - acima de R$ 1.500 → aumento de 10%

salario_atual = float(input("Digite o salário atual do funcionário: R$ "))


LIMITE_SALARIAL = 1500

# Estrutura condicional para calcular o aumento
if salario_atual <= LIMITE_SALARIAL:

    percentual_aumento = 0.15
    aumento = salario_atual * percentual_aumento
    novo_salario = salario_atual + aumento
else:

    percentual_aumento = 0.10
    aumento = salario_atual * percentual_aumento
    novo_salario = salario_atual + aumento


print(f"\nSalário atual: R$ {salario_atual:.2f}")
print(f"Aumento concedido: R$ {aumento:.2f} ({percentual_aumento*100:.0f}%)")
print(f"Novo salário: R$ {novo_salario:.2f}")
```

```python
#16. Elabore um algoritmo que leia 5 números inteiros e exiba-os em ordem inversa.

numeros = []


for i in range(5):
    numero = int(input(f"Digite o {i+1}º número inteiro: "))
    numeros.append(numero)

# Inverte a ordem dos elementos na lista
numeros.reverse()


print("Os números em ordem inversa são:", numeros)
```

```python
#17. Crie um algoritmo que leia as notas de 5 alunos, armazene em um vetor e calcule a média geral da turma.

notas_alunos = []
num_alunos = 5


print(f"Digite as {num_alunos} notas dos alunos:")
for i in range(num_alunos):
    while True: # Loop para garantir que a entrada é um número válido
        try:
            nota = float(input(f"Nota do aluno {i + 1}: "))
            # Opcional: verificar se a nota está entre 0 e 10
            if 0 <= nota <= 10:
                notas_alunos.append(nota)
                break # Sai do loop de verificação se a nota for válida
            else:
                print("Por favor, insira uma nota entre 0 e 10.")
        except ValueError:
            print("Entrada inválida. Por favor, insira um número.")

# 3. Calcule a soma das notas
soma_notas = sum(notas_alunos)

# 4. Calcule a média geral da turma
media_geral = soma_notas / len(notas_alunos)

# 5. Imprima o resultado
print(f"\nAs notas inseridas foram: {notas_alunos}")
print(f"A média geral da turma é: {media_geral:.2f}")
```

```python
#18. Desenvolva um algoritmo que leia 10 números e conte quantos são pares.

contagem_pares = 0

# Loop para ler 10 números
for i in range(10):
    # Pede ao usuário para digitar um número e o converte para inteiro
    numero = int(input("Digite um número: "))

    # Verifica se o número é par
    if numero % 2 == 0:
        # Se for par, incrementa o contador
        contagem_pares += 1

# Exibe o resultado
print(f"Foram lidos {contagem_pares} números pares.")
```

```python
#20. Implemente uma função que receba um número e retorne se ele é primo.

def eh_primo(numero):


    if numero < 2:
        return False
    for i in range(2, int(numero/2) + 1):
        if numero % i == 0:
            return False
    return True


numero_teste = 2
if eh_primo(numero_teste):
    print(f"{numero_teste} é um número primo.")
else:
    print(f"{numero_teste} não é um número primo.")

numero_teste = 80
if eh_primo(numero_teste):
    print(f"{numero_teste} é um número primo.")
else:
    print(f"{numero_teste} não é um número primo.")
```
